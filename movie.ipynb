# Step 1
Create a movie recommendation system by processing the data from "tmdb_5000_credits.csv" and "tmdb_5000_movies.csv". The final output should be three pickle files: `movies.pkl`, `movies_dicti.pkl`, and `similarity.pkl`.

## Step 2

Load the `tmdb_5000_credits.csv` and `tmdb_5000_movies.csv` files into pandas DataFrames.


**Reasoning**:
Import pandas and load the two CSV files into dataframes, then display the head and info of both dataframes to inspect the data as instructed.



```python
import pandas as pd

credits_df = pd.read_csv('/content/tmdb_5000_credits.csv')
movies_df = pd.read_csv('/content/tmdb_5000_movies.csv')

print("Credits DataFrame Head:")
display(credits_df.head())
print("\nCredits DataFrame Info:")
credits_df.info()

print("\nMovies DataFrame Head:")
display(movies_df.head())
print("\nMovies DataFrame Info:")
movies_df.info()
```

## Step 3

Merge the two DataFrames on the 'title' column.


**Reasoning**:
Merge the two dataframes on the 'title' column as instructed and then display the head and info of the merged dataframe.



```python
merged_df = movies_df.merge(credits_df, on='title')

print("Merged DataFrame Head:")
display(merged_df.head())
print("\nMerged DataFrame Info:")
merged_df.info()
```

## Step 4 Data preprocessing
Clean and preprocess the merged data, including handling missing values and selecting relevant features.


**Reasoning**:
Select relevant columns, handle missing values, and convert string representations of lists/dictionaries to actual lists/dictionaries.



```python
import ast

# Select relevant columns
merged_df = merged_df[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]

# Handle missing values in 'overview'
merged_df['overview'] = merged_df['overview'].fillna('')

# Drop rows with missing values in 'genres', 'keywords', 'cast', 'crew' if any
merged_df.dropna(subset=['genres', 'keywords', 'cast', 'crew'], inplace=True)

# Convert string representations of lists/dictionaries to actual lists/dictionaries
features = ['genres', 'keywords', 'cast', 'crew']
for feature in features:
    merged_df[feature] = merged_df[feature].apply(ast.literal_eval)

display(merged_df.head())
merged_df.info()
```

## Step 5 Feature extraction


Extract features like genres, keywords, cast, and crew.


**Reasoning**:
Define and apply functions to extract features from the 'genres', 'keywords', 'cast', and 'crew' columns, then preprocess these features and concatenate them into a 'tags' column. Finally, display the relevant parts of the dataframe to verify the results.



```python
def get_top_n_names(list_of_dicts, n=3):
    """Extracts the top N names from a list of dictionaries."""
    if not isinstance(list_of_dicts, list):
        return []
    names = [item['name'] for item in list_of_dicts if isinstance(item, dict) and 'name' in item]
    return names[:min(len(names), n)]

def get_director_name(list_of_dicts):
    """Extracts the director's name from a list of dictionaries in the 'crew' column."""
    if not isinstance(list_of_dicts, list):
        return ''
    for item in list_of_dicts:
        if isinstance(item, dict) and item.get('job') == 'Director':
            return item.get('name', '')
    return ''

# Apply the functions to extract features
merged_df['genres'] = merged_df['genres'].apply(get_top_n_names)
merged_df['keywords'] = merged_df['keywords'].apply(get_top_n_names)
merged_df['cast'] = merged_df['cast'].apply(get_top_n_names)
merged_df['director'] = merged_df['crew'].apply(get_director_name)

# Preprocess the extracted features
def preprocess_text(text):
    """Converts text to lowercase and removes spaces."""
    if isinstance(text, list):
        return [str(i).lower().replace(" ", "") for i in text]
    elif isinstance(text, str):
        return text.lower().replace(" ", "")
    else:
        return ""

merged_df['genres'] = merged_df['genres'].apply(preprocess_text)
merged_df['keywords'] = merged_df['keywords'].apply(preprocess_text)
merged_df['cast'] = merged_df['cast'].apply(preprocess_text)
merged_df['director'] = merged_df['director'].apply(preprocess_text)

# Concatenate features into 'tags' column
merged_df['tags'] = merged_df['overview'] + ' ' + merged_df['genres'].apply(lambda x: ' '.join(x)) + ' ' + \
                    merged_df['keywords'].apply(lambda x: ' '.join(x)) + ' ' + \
                    merged_df['cast'].apply(lambda x: ' '.join(x)) + ' ' + merged_df['director']

# Display results
display(merged_df[['title', 'genres', 'keywords', 'cast', 'director', 'tags']].head())
merged_df.info()
```

## Step 6


Compute a similarity matrix based on the extracted features.


**Reasoning**:
Compute the cosine similarity matrix based on the 'tags' column.



```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Instantiate TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)

# Fit and transform the 'tags' column
tfidf_matrix = tfidf_vectorizer.fit_transform(merged_df['tags'])

# Compute cosine similarity
similarity = cosine_similarity(tfidf_matrix)

print("Shape of TF-IDF matrix:", tfidf_matrix.shape)
print("Shape of similarity matrix:", similarity.shape)
```

## Step 7

Create a dictionary of movies for easy lookup.


**Reasoning**:
Create a dictionary to map movie titles to their indices in the merged_df DataFrame and print the first few items.



```python
movies_dicti = merged_df.reset_index().set_index('title')['index'].to_dict()

# Print the first few key-value pairs
print("First 5 items in movies_dicti:")
for i, (key, value) in enumerate(movies_dicti.items()):
    if i >= 5:
        break
    print(f"{key}: {value}")
```

## Step 8

Save the processed movie data, the movie dictionary, and the similarity matrix as pickle files (`movies.pkl`, `movies_dicti.pkl`, `similarity.pkl`).


**Reasoning**:
Save the processed dataframes and the similarity matrix to pickle files as instructed.



```python
import pickle

# Save merged_df to movies.pkl
with open('movies.pkl', 'wb') as f:
    pickle.dump(merged_df, f)

# Save movies_dicti to movies_dicti.pkl
with open('movies_dicti.pkl', 'wb') as f:
    pickle.dump(movies_dicti, f)

# Save similarity to similarity.pkl
with open('similarity.pkl', 'wb') as f:
    pickle.dump(similarity, f)

print("Files saved successfully: movies.pkl, movies_dicti.pkl, similarity.pkl")
```

```python
import pickle

with open('movies.pkl', 'rb') as f:
    movies = pickle.load(f)

with open('movies_dicti.pkl', 'rb') as f:
    movies_dicti = pickle.load(f)

with open('similarity.pkl', 'rb') as f:
    similarity = pickle.load(f)

print("Pickle files loaded successfully.")
```

## Step 4

Define a function that takes a movie title as input and returns a list of recommended movies using the loaded data.


**Reasoning**:
Define the `recommend` function to get movie recommendations based on similarity scores.



```python
def recommend(movie):
    """Recommends movies similar to the input movie based on cosine similarity."""
    if movie not in movies_dicti:
        return "Movie not found."

    movie_index = movies_dicti[movie]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    recommended_movies = []
    for i in movies_list:
        recommended_movies.append(movies.iloc[i[0]].title)
    return recommended_movies

# Test the function
print("Recommendations for Avatar:")
print(recommend('Avatar'))
print("\nRecommendations for a non-existent movie:")
print(recommend('NonExistentMovie'))
```

## Step 5

Create a Streamlit application with a dropdown menu to select a movie and a button to get recommendations.


**Reasoning**:
Create a Streamlit application with a dropdown menu to select a movie and a button to get recommendations as instructed.



```python
import streamlit as st

st.title('Movie Recommender System')

selected_movie_name = st.selectbox(
    'Select a movie:',
    (movies_dicti.keys()))

if st.button('Get Recommendation'):
    recommendations = recommend(selected_movie_name)
    st.write('Recommended Movies:')
    for movie in recommendations:
        st.write(movie)
```
